```python
import rasterio
import numpy as np
```

```python
image_file = "S2_2018_02.tif"

# Loading red and NIR bands
with rasterio.open(image_file) as src:
    band_red = src.read(3)

with rasterio.open(image_file) as src:
    band_nir = src.read(4)
```

```python
# Allowing division by zero value
numpy.seterr(divide='ignore', invalid='ignore')

# Calculating NDVI
ndvi = (band_nir.astype(float) - band_red.astype(float)) / (band_nir + band_red)
```

```python
# Setting the spatial characteristics of the output object to mirror the input
kwargs = src.meta
kwargs.update(
    dtype=rasterio.float32,
    count = 1)
```

```python
# Creating the output in depository
with rasterio.open('ndvi.tif', 'w', **kwargs) as dst:
        dst.write_band(1, ndvi.astype(rasterio.float32))
import matplotlib.pyplot as plt
plt.imsave("ndvi_assign3.jpeg", ndvi, cmap=plt.cm.Greens)
```

```python
#Plotting the final ndvi output
plt.imshow(ndvi, cmap="YlGn")
plt.colorbar()
plt.show()
```

```python
#Printing the important values
print('Max NDVI: {m}'.format(m=ndvi.max()))
print('Mean NDVI: {m}'.format(m=ndvi.mean()))
print('Median NDVI: {m}'.format(m=np.median(ndvi)))
print('Min NDVI: {m}'.format(m=ndvi.min()))
```
